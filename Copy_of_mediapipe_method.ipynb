```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize the image to 224x224
    transforms.ToTensor(),           # Convert the image to a PyTorch tensor
    transforms.Normalize(            # Normalize the image
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Main directory containing subdirectories with images for each gesture
train_path = "./train"

# List all subdirectories (assuming each folder contains images for one gesture)
gesture_folders = os.listdir(train_path)

for gesture_folder in gesture_folders:
    # Get the path to the current gesture folder
    gesture_folder_path = os.path.join(train_path, gesture_folder)

    # List all image files in the gesture folder
    image_files = os.listdir(gesture_folder_path)

    # Loop through each image file
    for image_file in image_files:
        # Get the path to the current image file
        image_path = os.path.join(gesture_folder_path, image_file)

        # Load the image using PIL
        image = Image.open(image_path)

        # Apply the transformations to the image
        transformed_image = transform(image)

        # Perform any further processing or save the transformed image as needed
        # For example, you can save the transformed image to a new directory
        save_path = f"./train-preprocess/{gesture_folder}"
        os.makedirs(save_path, exist_ok=True)
        save_name = f"{gesture_folder}_{image_file}"
        transformed_image_path = os.path.join(save_path, save_name)
        torchvision.utils.save_image(transformed_image, transformed_image_path)
```

```python
data = gesture_recognizer.Dataset.from_folder(
    dirname=train_path,
    hparams=gesture_recognizer.HandDataPreprocessingParams(
        shuffle=True)
)
train_data, validation_data = data.split(0.80)
```

```python
test_data = gesture_recognizer.Dataset.from_folder(
    dirname=test_path,
    hparams=gesture_recognizer.HandDataPreprocessingParams(
        shuffle=True,
        min_detection_confidence=0.5
    )
)
```

```python
hparams = gesture_recognizer.HParams(
    export_dir="exported_model",
    batch_size=2,
    epochs=15,
    learning_rate=0.001,
    shuffle=False)
```

```python
model_options = gesture_recognizer.ModelOptions(
    dropout_rate=0.05,
    layer_widths=[]
)

options = gesture_recognizer.GestureRecognizerOptions(hparams=hparams, model_options=model_options)
```

```python
model = gesture_recognizer.GestureRecognizer.create(
    train_data=train_data,
    validation_data=validation_data,
    options=options)
```

```python
loss, acc = model.evaluate(test_data, batch_size=2)
print(f"Test loss:{loss}, Test accuracy:{acc}")
```
